# Summary

This noteboook reoughly describe a robust procedure to localize cell protrusions from binary masks. The Following steps are performed to retrieve the protrusion locations:

- Extract individual Cells
- Compute Distance Map
- find Center by computing the maximum location of the distance map for each cell
- perform polar transformation around cell center
- find local maxima in polar cell outline profile

# Cell Extraction

Next, we extract each cell. Since we want the cell areas to be compatible, we choose a constant radius around the cell center that is able encompas all cells in the frame. Further, we choose a square FOV for each frame to avoid complexities durien polar coordinate transformations and back transformations. 

```python
radii = ap.get_all_radii([mask], [CLIST]) 
CELLS = [ap.extract_rad_frames([mask], [CLIST],int(radii.max()),i)[0] for i in np.unique(mask)[1:]]
```

# Polar coordinate transformation
Next Next we perform polar coordinate transformations and back transformations of the mask image, the morphological outline and compute the protrusion locations.

```python
POLS   = np.asarray([ap.pol_trans(i)   for i in CELLS])
PLIN   = np.asarray([ap.pol_outline(i) for i in POLS])
RI, AI = np.asarray([ap.inner_circ(i)  for i in PLIN]).T
RO, AO = np.asarray([ap.outer_max(i)   for i in PLIN]).T
TOKI   = np.asarray([ap.get_toki(i)    for i in PLIN])
```

There are various possibilities for computing the local maximas (peaks) e.g. comparing neighbors, wavelets, etc. However, simple usually works best...  

```python
i=4
GA = ndimage.gaussian_filter
mpoints  = signal.find_peaks_cwt(PLIN[i][:,0],np.arange(5,12))
peaks, _ = signal.find_peaks(GA(PLIN[i][:,0],10), distance=8,width=8)
plt.figure(figsize=(16,4))
plt.plot(PLIN[i][:,0],"-")
plt.plot(mpoints,PLIN[i][:,0][mpoints],"og")
plt.plot(peaks,PLIN[i][:,0][peaks],"o")
plt.xlim([0,len(PLIN[i][:,0])])
plt.show()

plt.figure(figsize=(16,4))
plt.plot(PLIN[i][:,1],PLIN[i][:,0],"-")
plt.plot(PLIN[i][:,1][mpoints],PLIN[i][:,0][mpoints],"og")
plt.plot(PLIN[i][:,1][peaks],PLIN[i][:,0][peaks],"o")
plt.xlim([0,PLIN[i][:,1].max()])
plt.show()
```

```python
def gdy(phi,amp):
    return np.sin(phi/180.*np.pi)*amp
def gdx(phi,amp):
    return np.cos(phi/180.*np.pi)*amp

def plot_pol_trafo(IMAGE,POL,PL,TK,ri,ai,ro,ao):
    dy,dx = IMAGE.shape[0]/2,IMAGE.shape[1]/2
    
    fig = plt.figure(figsize=(16,7))
    gs  = fig.add_gridspec(6,20)
    a   = fig.add_subplot(gs[:, :6])
    a.imshow(IMAGE,cmap="Greys",vmax=3)
    a.contour(IMAGE,colors="r", linewidths=0.5)
    a.plot(dx, dy,".k")
    a.plot(gdx(TK[:,1],TK[:,0])+dx,gdy(TK[:,1],TK[:,0])+dy,"o", color="cornflowerblue")
    
    ac = plt.Circle((dx, dy), 
                    ri,fill=False,color= "k",
                    linestyle="--")

    ad = plt.Circle((dx, dy), 
                     ro,fill=False,color= "k",
                     linestyle="--")

    ae = plt.Circle((dx, dy), 
                    ri*2**0.5,fill=False,color= "cornflowerblue",
                    linestyle="--")

    a.add_artist(ac)
    a.add_artist(ad)
    #a.add_artist(ae)

    a.arrow(dx, dy,
              gdx(ai,ri),gdy(ai,ri),
              length_includes_head=True,
              width=1, 
              head_width=6,
              head_length=12,
              color="k")


    a.arrow(dx, dy,
              gdx(ao,ro),gdy(ao,ro),
              length_includes_head=True,
              width=1, 
              head_width=6,
              head_length=12,
              color="k")

    #plt.ylim([100,380])
    #plt.xlim([35,445])

    a.set_title("Cell Morphology in Cartesian Coordinates", loc="left")
    a.set_ylabel(r"$y [\mu m]$")
    a.set_xlabel(r"$x [\mu m]$")

    b = fig.add_subplot(gs[:, 7:])
    b.imshow(POL,origin="lower",cmap="Greys",vmax=3)
    b.plot(PL[:,1],PL[:,0],"-r", markersize=1.3)
    b.plot([0,len(POL[0])],[ri,ri],"--k")
    b.plot([0,len(POL[0])],[ro,ro],"--", color="k")
    #b.plot([0,len(POL[0])],[ri*2**0.5,ri*2**0.5],"--", color="cornflowerblue")
    b.plot(TK[:,1],TK[:,0],"o", color="cornflowerblue")
    b.arrow(ai, 0, 0, ri,
              length_includes_head=True,
              width=1, 
              head_width=6,
              head_length=12,
              color="k")

    b.arrow(ao , 0, 0, ro  ,
              length_includes_head=True,
              width=1,
              head_width=6,
              head_length=12,
              color="k")

    b.set_title("Cell Morphology in Polar Coordinates", loc="left")
    b.set_ylabel(r"$d [\mu m]$")
    b.set_xlabel(r"$\phi [^{\circ}]$")
    b.set_xlim([0,360])
    b.set_ylim([0,180])
    plt.show()

for i in range(len(CELLS)):
    print("Cell index: ",i+1)
    plot_pol_trafo(CELLS[i],POLS[i],PLIN[i],TOKI[i],RI[i],AI[i],RO[i],AO[i])
```

```python
def gdy(phi,amp):
    return np.sin(phi/180.*np.pi)*amp
def gdx(phi,amp):
    return np.cos(phi/180.*np.pi)*amp

def plot_pol_trafo(IMAGE,POL,PL,TK,ri,ai,ro,ao):
    dy,dx = IMAGE.shape[0]/2,IMAGE.shape[1]/2
    
    fig = plt.figure(figsize=(10,7))
    gs  = fig.add_gridspec(12,20)
    a   = fig.add_subplot(gs[:6, :6])
    a.imshow(IMAGE,cmap="Greys",vmax=3)
    a.contour(IMAGE,colors="r", linewidths=0.5)
    a.plot(dx, dy,".k")
    a.plot(gdx(TK[:,1],TK[:,0])+dx,gdy(TK[:,1],TK[:,0])+dy,"o", color="cornflowerblue")
    
    ac = plt.Circle((dx, dy), 
                    ri,fill=False,color= "k",
                    linestyle="--")

    ad = plt.Circle((dx, dy), 
                     ro,fill=False,color= "k",
                     linestyle="--")

    ae = plt.Circle((dx, dy), 
                    ri*2**0.5,fill=False,color= "cornflowerblue",
                    linestyle="--")

    a.add_artist(ac)
    a.add_artist(ad)
    #a.add_artist(ae)

    a.arrow(dx, dy,
              gdx(ai,ri),gdy(ai,ri),
              length_includes_head=True,
              width=1, 
              head_width=10,
              head_length=20,
              color="k",
              zorder=20)


    a.arrow(dx, dy,
              gdx(ao,ro),gdy(ao,ro),
              length_includes_head=True,
              width=1, 
              head_width=10,
              head_length=20,
              color="k",
              zorder=21)
    
    plt.text(gdx(ai,ri)+dx-50,(gdy(ai,ri)+dy+10),r"$\vec{\epsilon}_1$",fontsize=20,color="k")
    plt.text(gdx(ao,ro)+dx-50,(gdy(ao,ro)+dy+20),r"$\vec{\epsilon}_2$",fontsize=20,color="k")
    #plt.ylim([100,380])
    #plt.xlim([35,445])

    a.set_title("A: Cell Morphology", loc="left")
    a.set_ylabel(r"$y [\mu m]$",labelpad=-2)
    a.set_xlabel(r"$x [\mu m]$",labelpad=-2)

    b = fig.add_subplot(gs[:6, 7:])
    b.imshow(POL,origin="lower",cmap="Greys",vmax=3)
    b.plot(PL[:,1],PL[:,0],"-r", markersize=1.3)
    b.plot([0,len(POL[0])],[ri,ri],"--k")
    b.plot([0,len(POL[0])],[ro,ro],"--", color="k")
    #b.plot([0,len(POL[0])],[ri*2**0.5,ri*2**0.5],"--", color="cornflowerblue")
    b.plot(TK[:,1],TK[:,0],"o", color="cornflowerblue")
    b.arrow(ai, 0, 0, ri,
              length_includes_head=True,
              width=1, 
              head_width=6,
              head_length=12,
              color="k",
              zorder=20)

    b.arrow(ao , 0, 0, ro  ,
              length_includes_head=True,
              width=1,
              head_width=6,
              head_length=12,
              color="k",
              zorder=21)
    
    plt.text(ai+6,ri/2-2,r"$\vec{\epsilon}_1$",fontsize=20,color="k")
    plt.text(ao+6,ro/2-2,r"$\vec{\epsilon}_2$",fontsize=20,color="k")
    b.set_title("B: Cell Morphology in Polar Coordinates", loc="left")
    b.set_ylabel(r"$d [\mu m]$",labelpad=-8)
    b.set_xlabel(r"$\phi [^{\circ}]$",labelpad=-2)
    b.set_xlim([0,360])
    
    c = fig.add_subplot(gs[7:, :])

    cxx = range(len(PL[:,0]))
    xscale = float(cxx[-1])/360.
    c.plot(cxx,PL[:,0][::-1],"-r", markersize=1.3)
    c.plot([0,cxx[-1]],[ri,ri],"--k")
    c.plot([0,cxx[-1]],[ro,ro],"--", color="k")
    c.set_title("C: Polar Distance vs. Angular Path Distance", loc="left")
    c.set_ylabel(r"$d [\mu m]$",labelpad=-2)
    c.set_xlabel(r"$S [\mu m]$",labelpad=-1.8)
    c.set_xlim([0,len(PL[:,0])])
    c.set_ylim([0,170])
    #plt.savefig("images/pol_map_toki.eps",bbox_inches='tight',quality=50,dpi=150)
    plt.show()

for i in range(len(CELLS))[:1]:
    print(i+1)
    plot_pol_trafo(CELLS[i],POLS[i],PLIN[i],TOKI[i],RI[i],AI[i],RO[i],AO[i])
```

```python
def plot_VT(CROPS,H=3,W=7):
    WW = 14
    HH = WW*H/W
    fig, axes = plt.subplots(H,W, figsize=(WW,HH), frameon = False)
    axs = axes.ravel()
    for i in range(len(CROPS)):
        axs[i].imshow(CROPS[i],cmap="Greys",vmax=3)
        axs[i].plot(CROPS[i].shape[1]//2,CROPS[i].shape[0]//2,".k")
        axs[i].set_xticks(()),axs[i].set_yticks(())
    fig.subplots_adjust(wspace=0)
    fig.subplots_adjust(hspace=0)
    plt.xticks(()),plt.yticks(())
    plt.show()
    
derot_cells = ap.derot_cells(CELLS,AO)
plot_VT(CELLS)
```

```python
LEN_RATIOS = RI/RO
derot_cells_align = ap.realign_rotcells(derot_cells)
X                 = np.arange(len(LEN_RATIOS))+1
dXP   = derot_cells_align.shape[1]/X.shape[0]
XP    = np.linspace(dXP/2,derot_cells_align.shape[1]-dXP/2,X.shape[0])
plt.figure(figsize=(10,5))
plt.subplot(211)
plt.title("Frame: "+str(1),loc="left")
plt.imshow(derot_cells_align,cmap="Blues",vmin=0.9,vmax=1.05)
for i in XP:
    plt.plot([i,i], [0,derot_cells_align.shape[0]],"--k",linewidth=0.8,alpha=0.5)
plt.xticks([]),plt.yticks([])
plt.subplot(212)
plt.bar(X,1./LEN_RATIOS)
for i in X:
    plt.plot([i,i], [0,10],"--k",linewidth=0.8,alpha=0.5)
plt.xticks(X)
plt.xlim([0.5,18.5])
plt.subplots_adjust(hspace=-0.2, wspace=-0.)
plt.xlabel("Cell index")
plt.ylabel(r"$| \vec{\epsilon}_2 |  / | \vec{\epsilon}_1 | $")
#plt.savefig("images/elon_cell.jpg",bbox_inches='tight',quality=95,dpi=150)
plt.ylim([0.,10])
plt.show()
```

```python
LR_smask            = np.argsort(LEN_RATIOS)
s_derot_cells       = derot_cells[LR_smask]
s_derot_cells_align = ap.realign_rotcells(s_derot_cells)
s_X                 = X[LR_smask]
s_LEN_RATIOS        = LEN_RATIOS[LR_smask]


plt.figure(figsize=(10,5))
plt.subplot(211)
plt.imshow(s_derot_cells_align,cmap="Blues",vmin=0.9,vmax=1.05)
for i in XP:
    plt.plot([i,i], [0,derot_cells_align.shape[0]],"--",linewidth=0.8,color="cornflowerblue")
plt.xticks([]),plt.yticks([])
plt.title(r"Frame: "+str(1)+r"l - ($|\epsilon| = | \vec{\epsilon}_1 |  / | \vec{\epsilon}_2 | $)",loc="left")
plt.subplot(212)
for i in X:
    plt.plot([i,i], [0,10],"--",linewidth=0.8,color="cornflowerblue")
plt.bar(X,s_LEN_RATIOS,tick_label=s_X,color="cornflowerblue")
plt.xlim([0.5,18.5])
plt.subplots_adjust(hspace=-0.2, wspace=0)
plt.xlabel("Cell index")
plt.ylabel(r"$| \vec{\epsilon}_1 |  / | \vec{\epsilon}_2 | $")
plt.ylim([0.,0.5])
#plt.savefig("images/elon_cell_sort.pdf",bbox_inches='tight')#,quality=95,dpi=150)
plt.show()
```
